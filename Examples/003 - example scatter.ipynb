### **Example 1**

We use the `scatter_chart` function to plot a series of random numbers.

```python
# Data
df = { 
        'x0': [1, 2, 3, 4, 5, 6],
        'y0': [1, 2, 3, 4, 5, 6],
        'x1': [7, 8, 9, 10, 11, 12],
        'y1': [7, 8, 9, 10, 11, 12],
        'x2': [13, 14, 15, 16, 17, 18],
        'y2': [13, 14, 15, 16, 17, 18],
        'x3': [19, 20, 21, 22, 23, 24],
        'y3': [19, 20, 21, 22, 23, 24],
     }

# Chart setup
chart_config = {
                'name': 'figure1-3-1',
                'width': 16., 
                'height': 8.0,
                'dots_per_inch': 600, 
                'extension': 'svg',
                'marker_size': 5,
                'color_map': ['red', 'blue', 'green', 'yellow'],
                'x_axis_label': 'Weight',
                'x_axis_size': 14,
                'y_axis_label': 'Height',
                'y_axis_size': 14,
                'axises_color': 'red', 
                'labels_size': 14,
                'labels_color': 'blue',
                'on_grid': False,
                'y_log': False,
                'x_log': False,
                'legend': ['data0', 'data1', 'data2', 'data3'],
                'legend_location': 'upper left',
                'size_legend': 12,
               }

# Call function
scatter_chart(dataset=df, plot_setup=chart_config)
```

### **Example 2**

```python
# Data
h = list(np.random.normal(loc=165, scale=10, size=2000))
w = list(np.random.gumbel(loc=65, scale=15, size=2000))
a = list(np.random.normal(loc=30, scale=3.0, size=2000))
df = {
        'x0': h,
        'y0': w,
        'colorbar': a
     }

# Chart setup
chart_config = {
                'name': 'figure1-3-1',
                'width': 16., 
                'height': 8.0,
                'dots_per_inch': 600, 
                'extension': 'svg',
                'marker_size': 100,
                'color_map': 'YlGn',
                'x_axis_label': 'Weight',
                'x_axis_size': 14,
                'y_axis_label': 'Height',
                'y_axis_size': 14,
                'axises_color': 'red', 
                'labels_size': 14,
                'labels_color': 'blue',
                'on_grid': False,
                'y_log': False,
                'x_log': False
               }

# Call function
scatter_chart(dataset=df, plot_setup=chart_config)
```
