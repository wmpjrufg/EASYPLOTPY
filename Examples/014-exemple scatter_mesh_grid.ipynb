### **Example 1**

We use the `scatter_mesh_grid` function to plot a series of random numbers.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = np.sin(np.sqrt(np.add.outer(x**2, y**2))) # Exemplo de funo para o mesh grid

# Data
df = {
    'x': y,
    'y': x,
    'z': z,  
    'x_points': np.random.uniform(-5, 5, 100),
    'y_points': np.random.uniform(-5, 5, 100),
}

# Chart setup
chart_config = {
                'name': 'scatter_mesh_grid_plot',
                'width': 8,
                'height': 6,
                'extension': 'png',
                'dots_per_inch': 100,
                'marker_size': 10,
                'color_map': 'viridis',
                'y_axis_label': 'Eixo Y',
                'y_axis_size': 12,
                'x_axis_label': 'Eixo X',
                'x_axis_size': 12,
                'labels_size': 10,
                'labels_color': 'black',
                'axises_color': 'gray',
                'on_grid': True,
                'y_log': False,
                'x_log': False,
            }

# Call function
scatter_mesh_grid(dataset=df, plot_setup=chart_config)
```
